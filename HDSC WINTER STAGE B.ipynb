```python
dff = df[['T2','T6']]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalised_df = pd.DataFrame(scaler.fit_transform(dff), columns = dff.columns)
features_df = normalised_df.drop(columns = ['T6'])
target_variable = normalised_df['T6']
```

```python
df = df.drop('date' , axis = 1)
df = df.drop('lights', axis = 1)
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalised_df = pd.DataFrame(scaler.fit_transform(df), columns = df.columns)
features_df = normalised_df.drop(columns = ['Appliances'])
target_variable = normalised_df['Appliances']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features_df, target_variable, test_size = 0.3, random_state =42)
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(x_train,y_train)
predicted_values = linear_model.predict(x_test)
```

## Mean Absolute Error (MAE) FOR LINEAR REGRESSION

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test,predicted_values)
round(mae,2)
```

## Residual Sum Of Square (RSS) FOR LINEAR REGRESSION

```python
rss = np.sum(np.square(y_test - predicted_values))
round(rss, 2)
```

## Mean Squared Error (MAE) FOR LINEAR REGRESSION

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test,predicted_values))
round(rmse,3)
```

### COEFFICIENT OF DETERMINATION (R-SQUARE)- LINEAR REGRESSION 

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, predicted_values)
round(r2_score, 2)
```

## RIDGE REGRESSION

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=0.4)
ridge_reg.fit(x_train, y_train)
predicted_values = ridge_reg.predict(x_test)
```

## Mean Squared Error (MAE) FOR RIDGE

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test,predicted_values))
round(rmse,3)
```

## LASSO REGRESSION

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train, y_train)
predicted_values = lasso_reg.predict(x_test)
```

## Mean Squared Error (MAE) FOR LASSO

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test,predicted_values))
round(rmse,3)
```

```python
def get_weights_df(model, feat, col_name):
    #this function returns the weight of every feature
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df
```
